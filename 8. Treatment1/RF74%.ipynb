# Data Mining/Cleaning

```python
test.isna().sum()
```

```python
train.isna().sum()
```

```python
train['Timestamp'] = pd.to_datetime(train['Timestamp'])
test['Timestamp'] = pd.to_datetime(test['Timestamp'])
```

```python
train['Year'] = train['Timestamp'].dt.year
train['Month'] = train['Timestamp'].dt.month
train['Day'] = train['Timestamp'].dt.day
train['Hours'] = train['Timestamp'].dt.hour
train['Minutes'] = train['Timestamp'].dt.minute
train['Seconds'] = train['Timestamp'].dt.second
train['Weekdays'] = train['Timestamp'].dt.weekday
train['Weeks'] = train['Timestamp'].dt.week
```

```python
test['Year'] = test['Timestamp'].dt.year
test['Month'] = test['Timestamp'].dt.month
test['Day'] = test['Timestamp'].dt.day
test['Hours'] = test['Timestamp'].dt.hour
test['Minutes'] = test['Timestamp'].dt.minute
test['Seconds'] = test['Timestamp'].dt.second
test['Weekdays'] = test['Timestamp'].dt.weekday
test['Weeks'] = test['Timestamp'].dt.week
```

```python
#drop comments column
Train = train.drop(['comments','Timestamp'], axis = 1)
Test = test.drop(['comments','Timestamp'], axis = 1)
```

```python
#filling values of missing columns with mode()[0]

#State column
Train['state'] = Train['state'].fillna(Train['state'].mode()[0])
Test['state'] = Test['state'].fillna(Test['state'].mode()[0])

#self_employed column
Train['self_employed'] = Train['self_employed'].fillna(Train['self_employed'].mode()[0])

#work_interfere column
Train['work_interfere'] = Train['work_interfere'].fillna(Train['work_interfere'].mode()[0])
Test['work_interfere'] = Test['work_interfere'].fillna(Test['work_interfere'].mode()[0])
```

```python
Train['Gender'].unique()
```

```python
Train['Gender'] = Train['Gender'].replace(['M', 'Male', 'male', 'm', 'Male-ish', 'maile', 'something kinda male?',
                                           'Cis Male', 'Mal', 'Male (CIS)', 'Make', 'male leaning androgynous',
                                           'Male ', 'Man', 'msle', 'Mail', 'cis male','Malr', 'Cis Man'],'MALE')
Test['Gender'] = Test['Gender'].replace(['M', 'Male', 'male', 'm', 'Male-ish', 'maile', 'something kinda male?',
                                         'Cis Male', 'Mal', 'Male (CIS)', 'Make', 'male leaning androgynous',
                                         'Male ', 'Man', 'msle', 'Mail', 'cis male','Malr', 'Cis Man'],'MALE')
```

```python
Train['Gender'] = Train['Gender'].replace(['Female', 'Trans woman', 'female', 'Trans-female', 'Cis Female', 'F',
                                           'Woman', 'f', 'queer/she/they', 'Femake', 'woman', 'fluid', 'Genderqueer', 
                                           'Female ', 'cis-female/femme', 'Female (trans)', 
                                           'queer', 'Female (cis)', 'femail'],'FEMALE')

Test['Gender'] = Test['Gender'].replace(['Female', 'Trans woman', 'female', 'Trans-female', 'Cis Female', 'F',
                                         'Woman', 'f', 'queer/she/they', 'Femake', 'woman', 'fluid', 'Genderqueer', 
                                         'Female ', 'cis-female/femme', 'Female (trans)', 
                                         'queer', 'Female (cis)', 'femail'],'FEMALE')
```

```python
Train['Gender'] = Train['Gender'].replace(['non-binary', 'Nah', 'All', 'Enby', 'Androgyne',
                                           'Agender', 'Guy (-ish) ^_^', 'Neuter', 'A little about you', 'p',
                                           'ostensibly male, unsure what that really means'],'OTHERS')

Test['Gender'] = Test['Gender'].replace(['non-binary', 'Nah', 'All', 'Enby', 'Androgyne',
                                         'Agender', 'Guy (-ish) ^_^', 'Neuter', 'A little about you', 'p',
                                         'ostensibly male, unsure what that really means'],'OTHERS')
```

```python
Train['Gender'].unique()
```

```python
Train['Age'].unique()
```

```python
Train['Age'] = Train['Age'].replace([99999999999, 1726, 29, 329, ], 31)
```

```python
Test['Age'].unique()
```

```python
Test['Age'] = Test['Age'].replace([-1],32)
```

```python
Train1 = Train
Test1 = Test
```

```python
for i in Train1.columns:
    if (Train1[i].dtypes == 'object'):
        Train1[i] = label.fit_transform(Train1[i])
        
for i in Test1.columns:
    if (Test1[i].dtypes == 'object'):
        Test1[i] = label.fit_transform(Test1[i])
```

```python
X = Train1.drop(['treatment'], axis = 1)
y = Train1[['treatment']]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3164)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log = log.fit(X_train, y_train)
y_pred1 = log.predict(X_test)
MSE = metrics.mean_squared_error(y_test,y_pred1)
print('MSE:     ', MSE)
rmse1 = np.sqrt(MSE)
print('rmse is:    ',rmse1)
f1_score1 = f1_score(y_test, y_pred1, average = 'weighted')
print("f1_score is:     ", f1_score1)
accuracy_score1 = accuracy_score(y_test, y_pred1)
print("accuracy_score is:     ", accuracy_score1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn = knn.fit(X_train, y_train)
y_pred6 = knn.predict(X_test)
MSE = metrics.mean_squared_error(y_test,y_pred6)
print('MSE:     ', MSE)
rmse6 = np.sqrt(MSE)
print('rmse is:    ',rmse6)
f1_score6 = f1_score(y_test, y_pred6, average = 'weighted')
print("f1_score is:     ", f1_score6)
accuracy_score6 = accuracy_score(y_test, y_pred6)
print("accuracy_score is:     ", accuracy_score6)
```
