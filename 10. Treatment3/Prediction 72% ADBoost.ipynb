```python
X = train.drop(['treatment'], axis = 1)
y = train[['treatment']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 500)
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor

lin = LinearRegression()
log = LogisticRegression()
Rd = Ridge()
Ls = Lasso()
KNN = KNeighborsRegressor()
NB = GaussianNB() 
SVM = SVR()
DT = DecisionTreeRegressor(criterion='mse', splitter='best', 
                           max_depth=5000, min_samples_split=2, min_samples_leaf=1, 
                           min_weight_fraction_leaf=0.0, max_features=None, 
                           random_state=None, max_leaf_nodes=None, 
                           min_impurity_decrease=0, min_impurity_split=None, 
                           ccp_alpha=0)
RF = RandomForestRegressor(n_estimators=500, criterion='mse', max_depth=None, 
                             min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, 
                             max_features='log2', max_leaf_nodes=None, min_impurity_decrease=0.0, 
                             min_impurity_split=None, bootstrap=True, oob_score=False, n_jobs=-1, 
                             random_state=None, verbose=0, warm_start=False, 
                             ccp_alpha=0.0, max_samples=None)

BR = BaggingRegressor(base_estimator = DT)
GBR = GradientBoostingRegressor()
xgboost=XGBRegressor()
```

```python
log.fit(X_train, y_train)
y_pred = log.predict(X_test)
MSE = metrics.mean_squared_error(y_test,y_pred)
print('MSE:     ', MSE)
rmse1 = np.sqrt(MSE)
print('rmse is:    ',rmse1)
f1_score1 = f1_score(y_test, y_pred, average = 'weighted')
print("f1_score is:     ", f1_score1)
accuracy_score1 = accuracy_score(y_test, y_pred)
print("accuracy_score is:     ", accuracy_score1)
```
